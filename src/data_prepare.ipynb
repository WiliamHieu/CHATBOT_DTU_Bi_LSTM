```python
# Lưu response bằng set 
import pandas as pd

# Load data from Excel
df = pd.read_excel('C:/Data Science/CHATBOT_DTU_Bi_LSTM/data/Chatbot_Dataset.xlsx', sheet_name='summary1')

data = []
current_tag = None
patterns = []
responses = set()  # Use a set to avoid duplicate responses

for index, row in df.iterrows():
    tag = row['Nhóm câu hỏi']
    question = row['Câu hỏi']
    answer = row['Câu trả lời']

    # New tag encountered
    if tag != current_tag:
        if current_tag is not None:
            # Save the previous tag's data
            data.append({
                'tag': current_tag,
                'patterns': patterns,
                'responses': list(responses)  # Convert set to list to serialize in JSON
            })
            patterns = []
            responses = set()
        current_tag = tag

    # Append the current question and store the response
    patterns.append(question)
    if pd.notna(answer):  # Check if the answer is not NaN
        responses.add(answer)  # Add to set, automatically avoiding duplicates

# Append the last tag's data
if current_tag is not None:
    data.append({
        'tag': current_tag,
        'patterns': patterns,
        'responses': list(responses)
    })

# Prepare the final dictionary
data_vf = {"intents": data}

# Save as JSON
import json
with open('output_file.json', 'w', encoding='utf-8') as f:
    json.dump(data_vf, f, ensure_ascii=False, indent=4)

data_vf
```

```python
# Tạo một danh sách các dictionary để lưu trữ thông tin về mỗi nhóm câu hỏi
data = []

# Duyệt qua từng dòng trong dataframe
current_tag = None
patterns = []
responses = []
for index, row in df.iterrows():
    tag = row['Nhóm câu hỏi']
    question = row['Câu hỏi']
    answer = row['Câu trả lời']
    
    # Nếu tag mới, lưu các patterns và responses của tag trước vào data
    if current_tag is not None and tag != current_tag:
        data.append({'tag': current_tag, 'patterns': patterns, 'responses': responses})
        patterns = []
        responses = []
    
    # Lưu patterns và responses vào danh sách tương ứng
    patterns.append(question)
    responses.append(answer)
    current_tag = tag

# Lưu nhóm câu hỏi cuối cùng vào data
data.append({'tag': current_tag, 'patterns': patterns, 'responses': responses})

for item in data:
    item['patterns'] = [x for x in item['patterns'] if not pd.isnull(x)]
    item['responses'] = [x for x in item['responses'] if not pd.isnull(x)]
```

```python
data_vf = {"intents": data}
```
